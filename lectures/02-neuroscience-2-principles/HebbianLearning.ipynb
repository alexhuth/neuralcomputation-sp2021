```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(1000) - x1
x = np.vstack([x1, x2])
```

```python
plt.plot(x[0], x[1], '.')
```

```python
## Hebb's Rule: dw = eta.y.x

w = np.array([0.01, 0])

iterations = 1000
eta = 3e-6

ws = np.zeros((iterations, 2))
ws[0] = w

for ii in range(1, iterations):
    for jj in range(x.shape[1]):
        y = w.dot(x[:,jj])
        delta_w = eta * y * x[:,jj]
        w += delta_w
    ws[ii] = w
```

```python
ws, ws.shape
```

```python
plt.plot(ws[:,0])
plt.plot(ws[:,1])
plt.xlabel("Epochs")
```

```python
plt.plot(x[0], x[1], '.')
plt.plot(ws[:,0], ws[:,1], '.-')
```

# Oja's rule

```python
## Oja's rule: dw = (y.x - y^2.w)

w = np.array([0.01, 0])

iterations = 1000
eta = 2e-5

ws = np.zeros((iterations, 2))
ws[0] = w

for ii in range(1, iterations):
    for jj in range(x.shape[1]):
        y = w.dot(x[:,jj])
        delta_w = eta * (y * x[:,jj] - y**2 * w)
        w += delta_w
    ws[ii] = w

```

```python
plt.plot(ws[:,0])
plt.plot(ws[:,1])
```

```python
plt.plot(x[0], x[1], '.')
plt.plot(ws[:,0], ws[:,1], '.-')
```
